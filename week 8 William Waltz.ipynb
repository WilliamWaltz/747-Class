```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
url = 'https://archive.ics.uci.edu/dataset/2/adult'
df = pd.read_csv('adult.data', header=None)
columns = ['age', 'workclass', 'fnlwgt', 'education', 'education_num', 'martial_status', 'occupation',
           'relationship', 'race', 'gender', 'capital_gain', 'captial_loss', 'hours_per_week', 'native_country', 'income']
df.columns = columns
print(df.head())
print(df.info())
print(df.describe())
print(df.isnull().sum())
print(df.duplicated().sum())
df.replace('?', np.nan, inplace=True)
df.drop_duplicates(inplace=True)
print(df['age'].quantile(0.96))
print(df['hours_per_week'].quantile(0.96))
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3 - Q1
df= df[~((df['age'] < (Q1 - 1.2 * IQR)) | (df['age'] > (Q3 + 1.5 * IQR)))]
print(df.groupby('income')['hours_per_week'].mean())
print(df.groupby('education')['age'].mean())
plt.hist(df['age'],bins=10)
plt.show()
df.to_csv('Cleaned_Adult_Data.csv', index=False)
          
```
