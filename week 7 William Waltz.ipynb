```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
titanic = sns.load_dataset('titanic')
titanic = titanic.dropna(subset=['age', 'fare'])
z_score_age = stats.zscore(titanic['age'])
z_score_fare = stats.zscore(titanic['fare'])
sns.scatterplot(x='age', y='fare', data=titanic)
outlier_age = titanic[(z_score_age > 2) | (z_score_age <-2)]
outlier_fare = titanic[(z_score_fare > 2) | (z_score_fare <-2)]
plt.scatter(outlier_age['age'], outlier_age['fare'], color='green')
plt.scatter(outlier_fare['age'], outlier_fare['fare'], color='red')
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
titanic = sns.load_dataset('titanic')
titanic = titanic.dropna(subset=['age', 'fare', 'pclass'])
corr = titanic[['age', 'fare', 'pclass', 'survived']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', square=True)
plt.show()
```
